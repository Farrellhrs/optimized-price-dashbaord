```python
import pandas as pd
```

```python
df = pd.read_csv('dummy_forecasting_data_100weeks1.csv')
df.head()
```

```python
# Assign a fixed normal_price for each category
fixed_prices = df.groupby('category')['normal_price'].first().to_dict()
# Only update normal_price for rows where it differs from the fixed price
df['normal_price'] = df.apply(
    lambda row: fixed_prices[row['category']] if row['normal_price'] != fixed_prices[row['category']] else row['normal_price'],
    axis=1
)
```

```python
# Check if all normal_price values are the same within one category, e.g., 'Biskuit'
category = 'Sirup'
prices = df[df['category'] == category]['normal_price'].unique()
print(f"Unique normal_price values in category '{category}': {prices}")
print("All prices are the same:" if len(prices) == 1 else "Prices vary within the category.")
```
